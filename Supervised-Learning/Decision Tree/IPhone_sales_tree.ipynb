```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/iphone_purchase_records.csv")
df.head()
```

```python
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
df.head()
```

```python
X=df.drop('Purchase Iphone',axis=1)
y=df['Purchase Iphone']
X_train,X_test, y_train , y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#Creating Decision Tree model
dtm = DecisionTreeClassifier(random_state=42)
dtm.fit(X_train,y_train)

y_pred = dtm.predict(X_test)
```

```python
#Accuracy
acc = accuracy_score(y_test,y_pred)
print(f"Accuracy: {acc}")

#Classification Report
print("\nClassification Report:")
print(classification_report(y_test,y_pred, target_names = ['No', 'Yes']))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred))
```

```python
#plot decision tree
plt.figure(figsize=(12,12))
plot_tree(dtm,feature_names = X.columns,class_names=['No', 'Yes'],filled=True)
plt.title("Decision Tree")
plt.show()


#depth of tree
print(f"Depth of Tree: {dtm.get_depth()}")
```
