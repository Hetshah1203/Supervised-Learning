```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.drop(['petal_length','petal_width','species'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(iris,test_size=0.2)
x_train=train.drop('sepal_length',axis=1)
y_train=train['sepal_length']

x_test=test.drop('sepal_length',axis=1)
y_test=test['sepal_length']
```

```python
from sklearn import neighbors
from sklearn.metrics import mean_squared_error
from math import sqrt

model = neighbors.KNeighborsRegressor(n_neighbors=3)
model.fit(x_train,y_train)
pred = model.predict(x_test)
error = sqrt(mean_squared_error(y_test,pred))
error
```

```python
rmse_val = []
k_list=[]
for K in range(50):
  K=K+1
  model = neighbors.KNeighborsRegressor(n_neighbors=K)

  model.fit(x_train,y_train)
  pred = model.predict(x_test)
  error = sqrt(mean_squared_error(y_test,pred))
  rmse_val.append(error)
  k_list.append(K)
  print('RMSE value for k = ' , K , 'is:', error)
```

```python
best_rmse_score = min(rmse_val)
best_k = k_list[rmse_val.index(best_rmse_score)]
print("Optimal value for k = ", best_k , "with least RMSE =", best_rmse_score)
```

```python
plt.plot(k_list,rmse_val)
plt.xlabel('K')
plt.ylabel('RMSE')
```
